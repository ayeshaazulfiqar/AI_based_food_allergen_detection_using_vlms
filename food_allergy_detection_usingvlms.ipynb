```python
from google import genai
from PIL import Image
import pathlib

client = genai.Client(api_key="AIzaSyAXxVjKM0aULcEanNKfA7Vs0HxUVxzjE-U")

def detect_allergens(image_path=None, ingredients_text=None):
    
    prompt = """
    You are an expert food allergen detection system.
    
    Analyze the provided food information and return ONLY a JSON response in this exact format:
    {
        "dish_name": "Name of the dish",
        "description": "Brief one line description of the dish",
        "detected_allergens": ["Milk", "Egg"],
        "allergen_risk": {
            "Milk": 85,
            "Egg": 90,
            "Peanut": 0,
            "Tree_Nuts": 0,
            "Wheat_Gluten": 75,
            "Soy": 0,
            "Fish": 0,
            "Shellfish": 0,
            "Sesame": 0,
            "Mustard": 10,
            "Celery": 0,
            "Molluscs": 0
        },
        "confidence": "High/Medium/Low",
        "warning": "Any important allergen warning here or null"
    }
    
    Rules:
    - allergen_risk values are 0-100 (probability percentage)
    - detected_allergens only lists allergens with risk > 50
    - Be thorough and consider hidden allergens (e.g. soy in sauces)
    - Return ONLY the JSON, no extra text
    """
    
    contents = []
    
    # Add image if provided
    if image_path:
        img = Image.open(image_path)
        contents.append(img)
    
    # Add ingredients text if provided
    if ingredients_text:
        contents.append(f"Ingredients: {ingredients_text}")
    
    contents.append(prompt)
    
    response = client.models.generate_content(
        model="gemini-flash-lite-latest",
        contents=contents
    )
    
    return response.text

# Test with text only
result = detect_allergens(ingredients_text="eggs, whole milk, all purpose flour, butter, sugar, vanilla extract")
print(result)
```
